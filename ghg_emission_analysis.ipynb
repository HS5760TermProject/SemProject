```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore")  # Ignore warnings for clarity

# Load the data
# Assuming `elec_prod_data` is a DataFrame with 'DATE' and 'Value' columns

# Convert 'DATE' column to datetime format and set it as the index
elec_prod_data['DATE'] = pd.to_datetime(elec_prod_data['DATE'], format='%d-%m-%Y')
elec_prod_data.set_index('DATE', inplace=True)

# Plot the historical data
plt.figure(figsize=(12, 6))
plt.plot(elec_prod_data.index, elec_prod_data['Value'], color='b')
plt.title('Electricity Production Over Time')
plt.xlabel('Year')
plt.ylabel('Electricity Production')
plt.grid(True)
plt.show()

# Build and fit the ARIMA model
# ARIMA(p, d, q) parameters - p is the lag order, d is the degree of differencing, and q is the order of moving average
model = ARIMA(elec_prod_data['Value'], order=(5, 1, 0))  # Adjust order as needed
fitted_model = model.fit()

# Forecast for the next 5 years (you can adjust the periods)
forecast_periods = 5 * 365  # 5 years in days (assuming daily data)
forecast = fitted_model.forecast(steps=forecast_periods)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(elec_prod_data.index, elec_prod_data['Value'], label='Historical Data', color='b')
plt.plot(pd.date_range(elec_prod_data.index[-1], periods=forecast_periods, freq='D'), forecast, color='r', label='Forecast')
plt.title('Electricity Production Forecast')
plt.xlabel('Year')
plt.ylabel('Electricity Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore")  # Ignore warnings for clarity
elec_prod_data = pd.read_csv('Electric_Production.csv')

# Load and preprocess the electricity production data
# Assuming `elec_prod_data` is a DataFrame with 'DATE' and 'Value' columns
elec_prod_data['DATE'] = pd.to_datetime(elec_prod_data['DATE'], format='%d-%m-%Y')
elec_prod_data.set_index('DATE', inplace=True)

# Resample the data to annual average
elec_prod_data_yearly = elec_prod_data.resample('Y').mean()
elec_prod_data_yearly['year'] = elec_prod_data_yearly.index.year

# Load CO2 emissions data and select relevant columns
# Assuming `overall_co2_emissions` has 'year' and 'CO2 emissions' columns
# Merge the yearly electricity production with CO2 emissions on 'year'
merged_data_2 = pd.merge(overall_co2_emissions, elec_prod_data_yearly, on='year', how='inner')

# Plotting the values for the common years
plt.figure(figsize=(10, 6))
plt.plot(merged_data_2['Value'], merged_data_2['CO2 emissions'], label='Electricity Production vs CO2 Emissions', marker='o', color='blue')
plt.title('Electricity Production vs CO2 Emissions for Common Years')
plt.ylabel('CO₂ Emissions')
plt.xlabel('Electricity Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the 'Value' (Electricity Production) and 'primary_energy_consumption' columns
merged_data_2['Normalized_Value'] = scaler.fit_transform(merged_data_2[['Value']])
merged_data['Normalized_primary_energy_consumption'] = scaler.fit_transform(merged_data[['primary_energy_consumption']])
merged_data_2['Normalized_CO2_emissions'] = scaler.fit_transform(merged_data_2[['CO2 emissions']])
merged_data['Normalized_CO2_emissions'] = scaler.fit_transform(merged_data[['CO2 emissions']])

# Plotting the normalized values
plt.figure(figsize=(10, 6))

# Plot Electricity Production vs CO₂ Emissions
plt.plot(merged_data_2['Normalized_CO2_emissions'], merged_data_2['Normalized_Value'], label='Normalized Electricity Production vs CO₂ Emissions', marker='o', color='blue')

# Plot Electricity Consumption vs CO₂ Emissions
plt.plot(merged_data['Normalized_CO2_emissions'], merged_data['Normalized_primary_energy_consumption'], label='Normalized Electricity Consumption vs CO₂ Emissions', marker='s', color='orange')

plt.title('Normalized Electricity Production & Consumption vs CO₂ Emissions')
plt.xlabel('Normalized CO₂ Emissions')
plt.ylabel('Normalized Electricity')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure data is sorted by Temperature to show increasing trend properly
merged_data_3 = merged_data_3.sort_values(by='Temperature').reset_index(drop=True)

# Filter out rows where Temperature or primary energy consumption decreases
filtered_data = merged_data_3[(merged_data_3['Temperature'].diff() >= 0) & 
                              (merged_data_3['primary_energy_consumption'].diff() >= 0)]

# Plot the filtered data
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Temperature'], filtered_data['primary_energy_consumption'], color='blue', label='Filtered Data', marker='s')

plt.xlabel('Temperature')
plt.ylabel('Primary Electricity Consumption')
plt.legend()
plt.title('Temperature vs Primary Electricity Consumption')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Starting with the original data
temperature = merged_data_3['Temperature']
energy_consumption = merged_data_3['primary_energy_consumption']

# Filtering data to keep only the rows where both columns are strictly increasing
increasing_indices = (temperature.diff() > 0) & (energy_consumption.diff() > 0)
filtered_data = merged_data_3[increasing_indices]

# Plotting the filtered data
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Temperature'], filtered_data['primary_energy_consumption'], label='Filtered Increasing Data', color='blue')

plt.xlabel('Temperature')
plt.ylabel('Primary Energy Consumption')
plt.legend()
plt.title('Filtered Increasing Trend in Temperature vs Primary Energy Consumption')
plt.show()
```
